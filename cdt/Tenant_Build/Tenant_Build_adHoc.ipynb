```python
import subprocess as sp
from io import StringIO
import pandas as pd
tableData=""
```

```python
query = """
    SELECT *
    FROM swh.tenant_build
    WHERE wd_event_date IS NOT NULL
        AND
            build_type = 'Foundation Tenant Build'
    ORDER BY wd_event_date ASC
"""
```

```python
tableData = sp.check_output(
            f"""pharos sql run --sql "{query}" | jq -r '.result.data'""",
            shell=True,
        ).decode("utf-8")
```

```python
tableData = pd.read_csv(StringIO(tableData))
```
